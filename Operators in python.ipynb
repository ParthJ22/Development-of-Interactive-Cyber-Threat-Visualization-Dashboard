# 4.Assignment operators

```python
* Assignment operators are used in Python to assign values to variables
 Eg: a = 10 
Augmented Assignment Operator:
* It combines assignment operator with other logical operators like addition, multiplication etc.
* This operators that allows to write shortcut assignment statement
```

```python
i = 18
print(i)
i+=8   # i = i+8
print(i)
```

```python
i = 14
i-=8  # i = i-8
print(i)
```

```python
i = 12
i*=8    # i = i*8
print(i)
```

```python
i = 16
i/=8   # i = i/8
print(i)
```

```python
i = 16
i//=8     # i = i//8
print(i)
```

```python
i = 14
i%=8    # i = i%8
print(i)
```

# 5.Membership operators

```python
x = 121
list1 = [12,21,32,23,44,'Deccan']
print(x in list1)     
print(x not in list1) 
print(12 in list1)     
print(12 not in list1) 
```

```python
string = 'data science'
print('data' in string)  
print('Data' in string)  
```

```python
string = 'data science'
substring1 = 'Data'
substring2 = 'Python'

substring1 in string and substring2 in string # False and False
```

```python
string = 'Python and Data science'
substring1 = 'Data'
substring2 = 'Python'

substring1 in string and substring2 in string # True and True
```
