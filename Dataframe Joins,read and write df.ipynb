# Types of joins in dataframes

## 1. merge

```python
>> It is used to join 2 tables
>> To join 2 tables there must be a common column in 2 tables

Syntax >> pd.merge(left_df,right_df,left_on,right_on,how)
```

```python
# Read the emp_2.csv
left_df = pd.read_csv('emp_2.csv')
left_df
```

```python
# Read the dept.csv
right_df = pd.read_csv('dept.csv')
right_df
```

```python
# inner join  # pd.merge(left_df,right_df,left_on,right_on,how)
pd.merge(left_df,right_df,left_on='did',right_on='dept_id',how='inner')  # how = 'inner >> default value of how
```

```python
# left join
pd.merge(left_df,right_df,left_on='did',right_on='dept_id',how='left') 
```

```python
# right join
pd.merge(left_df,right_df,left_on='did',right_on='dept_id',how='right')
```

```python
# outer join
pd.merge(left_df,right_df,left_on='did',right_on='dept_id',how='outer')
```

```python
# Join the 2 df on dept_id column
pd.merge(left_df,right_df,on='dept_id',how='outer')
```

# 2. concat

```python
>> It is used to find union of 2 dataframes

syntax >> pd.concat([df1,df2,df3 ...],axis=0,join)
```

```python
# Read the 50_Startups.csv
df = pd.read_csv('50_Startups.csv')
df
```

```python
# First 5 rows
A = df[:5]
A
```

```python
# Next 5 rows
B = df[5:10]
B
```

```python
# Apply concat function
pd.concat([A,B])  # axis=0 >> default
```

```python
# Apply concat function
pd.concat([A,B],axis=1)  
```
