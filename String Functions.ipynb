# 24. center

```python
It returns a centered string of length width

string.center(width, fillchar=' ')
```

```python
# Total length=10
string = 'python'
string.center(10)
```

```python
# Total length=7
string = 'python'
string.center(7)
```

```python
# Total length=80 and '+' as fillparameter
string = 'data science'
string.center(80,'+')
```

```python
# Total length=80 and '+' as fillparameter
string = 'data science'
string.center(80,'*')
```
