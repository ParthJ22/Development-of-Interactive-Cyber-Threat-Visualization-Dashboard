# 2.Type of Variables

```python
1. global variables  >> Variables which are declared outside the function
2. local variables   >> Variables which are declared inside the function

We can access global variable from outside the function and inside the function
local >> We can not access local variables from outside the function
```

```python
def function1():  # Defining a fucntion
    print('python')
function1()          # Calling a function
```

```python
name = 'Rohit sharma'     # Global variable
def function1():  
    x = 100               # Local variable
    
function1()   
```

```python
name = 'Rohit sharma'     # Global variable
def function1():  
    x = 100               # Local variable
    print(name)
    
function1()   
```

```python
name = 'Rohit sharma'     # Global variable
def function1():  
    x = 100               # Local variable
    print(name)
    
function1()   
print(name)
```

```python
name = 'Rohit sharma'     # Global variable
def function1():  
    x = 100               # Local variable
    print(name)
    print(x)
    
function1()   
print(name)
```

```python
name = 'Rohit sharma'     # Global variable
def function1():  
    l = 100               # Local variable
    print(name)
    print(l)
    
function1()   
print(name)
print(l)
```

```python
name = 'Rohit sharma'     # Global variable
def function1(): 
    global l
    l = 100               # Local variable
    print(name)
    print(l)
    
function1()   
print(name)
print(l)
```
