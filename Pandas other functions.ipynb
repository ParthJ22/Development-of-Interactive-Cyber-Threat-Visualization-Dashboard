# 1.value_counts

```python
>> It returns a series of unique value counts

syntax>> df[categorical_column_name].value_counts()

Note - Dont use value_counts() function on any numerical column. Use this function only on
       categorical column.
```

```python
# Find value counts of STATE column
df['STATE'].value_counts()
```

# 7. insert

```python
>> This function is used to insert column at a particular index in a dataframe.

syntax >> df.insert(index_number,column_name,value,allow_duplicates=False)
```

```python
array = np.random.randint(1,10,size=(5,5))
array
```

```python
# create a dataframe having 5 columns and 5 rows
df = pd.DataFrame(array,columns=['A','B','C','D','E'])
df
```

```python
# Add a column having 1000 value
df.insert(2,'Add1000',1000)
df
```

```python
# Add a column with python as value
df.insert(4,'Add_python','python')
df
```

```python
# Add a column with same column name as above and  with python as value
df.insert(5,'Add_python','python')   # allow_duplicates = False
df
```

```python
# use allow_duplicates=True
df.insert(5,'Add_python','python',allow_duplicates=True)   # allow_duplicates = True
df
```

```python
# Add square column
df.insert(1,'Square_A',df['A']**2)
df
```

```python
# Add cube column
df.insert(3,'Cube_B',df['B']**3)
df
```

```python
# Add a list as a value
# df.insert(index_number,column_name,value,allow_duplicates=False)

df.insert(6,'F',[10,20,30,40,50])
df
```

```python
# Add a array as a value
df.insert(7,'G',np.array(['python','data',200,300,500]))
df
```

# 8. rename

```python
>> This function is used to rename the columns or rows in a dataframe

Syntax:
df.rename({oldname1:newname1,oldname2:newname2},axis=0)  >> To rename 1 or 2 row names
df.rename({oldname1:newname1,oldname2:newname2},axis=1)  >> To rename 1 or 2 column names

df.columns = column_names_list   >> To rename all column names
df.index   = row_names_list      >> To rename all row names
```
