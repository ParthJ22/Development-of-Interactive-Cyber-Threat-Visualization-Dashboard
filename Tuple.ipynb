# Tuple operations

```python
# use '+' operator on below variables
city_names = ('Mumbai','PUNE',"Delhi")
state_names = ('Maharashtra','Karnataka','Gujrat')

city_names + state_names
```

```python
# using '*' operator 
city_names = ('Mumbai','PUNE',"Delhi")

print(city_names*3)
```

# Access item using for loop

```python
# using for loop access all items in tuple
tuple1 = ('Mumbai', 'PUNE', 'Delhi', 'Maharashtra', 'Karnataka', 'Gujrat')

for item in tuple1:
    print(item)
```

```python
# use enumerate function
tuple1 = ('Mumbai', 'PUNE', 'Delhi', 'Maharashtra', 'Karnataka', 'Gujrat')

for index,item in enumerate(tuple1):
    print(index,item)
```
