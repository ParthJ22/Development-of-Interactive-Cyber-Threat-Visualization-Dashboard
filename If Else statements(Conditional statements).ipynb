# WAP to calculate tax based on criteria:

```python
1. Agro products --> 0% tax
2. Clothing --> 10% tax
3. Furniture --> 11% tax
4. Electronics --> 28% tax
5. Alchoholic beverage --> 150% tax
6. Cosmetics --> 20% tax
7. Other --> 30%
```

```python
item_price = int(input("Enter the price of item: "))
item_type = int(input("Select item type:\n1. Agro products --> 0% tax\n2. Clothing --> 10% tax\n3. Furniture --> 11% tax\n4. Electronics --> 28% tax\n5. Alchoholic beverage --> 150% tax\n6. Cosmetics --> 20% tax\n7. Other --> 30%\n"))

if(item_type == 1):
    tax = 0
elif(item_type == 2):
    tax = item_price*0.1
elif(item_type == 3):
    tax = item_price*0.11
elif(item_type == 4):
    tax = item_price*0.28
elif(item_type == 5):
    tax = item_price*1.5
elif(item_type == 6):
    tax = item_price*0.2
else:
    tax = item_price*0.3
    
print("Total Tax to be paid:",tax)
print("Item value after tax:",item_price+tax)
```
