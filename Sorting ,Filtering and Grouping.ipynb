# 1.Sorting

## sort_values function

```python
>> This function is used to sort the dataframe based on a column values
    ascending(lower-higher)(a-z)
    descending(higher-lower)(z-a)

syntax>> df.sort_values(column_name,ascending=True)  # Sort the dataframe in ascending order
         df.sort_values(column_list,ascending=True)
    
         df.sort_values(column_name,ascending=False) # Sort the dataframe in descending order

ascending = True  >> ascending order (default)
ascending = False >> descending order
```

```python
# Sort the df based on RND column in ascending order
df.sort_values('RND',inplace=True)     # ascending = True
```

```python
df
```

```python
# Reset the index
df.index = range(0,50)
df
```

```python
# Sort the df based on RND column in descending order
df.sort_values('RND',ascending=False)     # ascending = True
```

```python
# Sort the df based on MKT column in descending order
df.sort_values('MKT',ascending=False)
```

```python
# Sort the df based on STATE,PROFIT column in ascending order
df.sort_values(['STATE','PROFIT'])
```

```python
# Sort the df based on STATE,PROFIT column in descending order
df.sort_values(['STATE','PROFIT'],ascending=False)
```

```python
df2 = pd.read_csv('Credit.csv')
df2
```

```python
df2.drop(['Unnamed: 0','ID'],axis=1,inplace=True)
df2
```

```python
#4.Find out Ethinicity wise count of candidates
df2.groupby('Ethnicity').count()
```

```python
#5.Find out total number of married and unmmaried candidates
df2.groupby('Married')['Income'].count()
```
